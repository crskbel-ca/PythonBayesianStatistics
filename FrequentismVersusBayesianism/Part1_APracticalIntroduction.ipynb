```python
# http://jakevdp.github.io/blog/2014/03/11/frequentism-
# and-bayesianism-a-practical-intro/


# Generating simple photon count data
import numpy as np
from scipy import stats

from IPython.display import Image
```

```python
np.random.seed(1)
F_true = 1000
N = 50 # number of measurements
F = stats.poisson(F_true).rvs(N)  # N measurements of flux
e = np.sqrt(F) # errors on Poisson counts estimated with sqrt rule. 
print("F = {0},\n\ne = {1}".format(F, e))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
#ax.set_facecolor('white') # already the default
plt.gcf().set_facecolor('white') # axes bg color 
ax.errorbar(F, np.arange(N), xerr=e, fmt='ok', ecolor='magenta', alpha=0.7)
ax.vlines([F_true], 0, N, linewidth=5, alpha=0.2)
ax.set_xlabel("Flux")
ax.set_ylabel("Measurement Number")
```

```python
# GOAL QUESTION: Given our measurements and errors what is the best
# estimate of the true flux?
```

```python
# value of trueFlux such that likelihood is maximized
# results in wi = 1/ei^2
w = 1. / e**2
print(""" 
      F_true = {0}
      F_est = {1:.0f} +/- {2:.0f} (based on {3} measurements)
      """.format(F_true, 
                 sum(w * F) / sum(w),
                 sum(w)**-0.5,  # error estimate of 0.4%
                 N))
# w.sum()**(-0.5) # this is the stdeviation.
```

```python
# Conclusion: for 50 measurements of the flux our estimate
# F_est is around 998 and has error (stdev) of 0.4%.
```

```python
## BAYESIAN APPROACH TO PHOTON COUNTS

# want to find: P(F_true | D) --- frequentism statistics says this has no
# meaning because probabilities have no meaning for model parameters.

# P(F_true | D) = P(F_true) * P(D | F_true) / P(D)

# P(F_true) = model prior, better to make this choice explicit and include
# some subjectivity rather than skirting the question. 
# P(D) = data probability, normalization term. 
# P(D | F_true) = likelihood
# P(F_true | D) = posterior, probability of model params given data.

# Goal: use emcee (MCMC sampling) to generate data points drawn from 
# posterior distribution and use those to determine the answer.

Image(filename="images/bayesformula.png")
```

```python
# P(F_true)
def logPrior(theta): # theta = H = hypothesis = F_true 
    return 1 # flat prior 

# P(D | F_true)
def logLikelihood(theta, F, e): # theta = Hypo
    return -0.5 * np.sum(np.log(2 * np.pi * e ** 2)
                         + (F - theta[0]) ** 2 / e ** 2)

# P(F_true | D)
def logPosterior(theta, F, e):
    # adding means multiplication on real number line. 
    return logPrior(theta) + logLikelihood(theta, F, e)
```

```python
# Goal of MCMC: generate sample points from the posterior. Then
# after we use those points to get probability we seek. 

## generating some random starting guesses
ndim = 1 # num parameters in model 
nwalkers = 50 # number of MCMC walkers
nburn = 1000 # burn in period to let chains stabilize
nsteps = 2000 # number of MCMC steps to take


# start at random locations between 0 and 20
startingGuesses = 2000 * np.random.rand(nwalkers, ndim)

import emcee
sampler = emcee.EnsembleSampler(nwalkers, ndim, logPosterior, args=[F, e])
sampler.run_mcmc(startingGuesses, nsteps)

sample = sampler.chain # shape = (nwalkers, nsteps, ndim)
sample = sampler.chain[:, nburn:, :].ravel() # discard burn-in points
```

```python
# plot histogram of the sample (sample contains 50000 points drawn from
# the posterior). 
plt.gcf().set_facecolor('white') # axes bg color 
plt.hist(sample, bins=50, histtype="stepfilled", alpha=0.5, normed=True)

# plot a best-fit Gaussian
F_fit = np.linspace(975, 1025)
pdf = stats.norm(np.mean(sample), np.std(sample)).pdf(F_fit)

plt.plot(F_fit, pdf, "-k")
plt.xlabel("F")
plt.ylabel("P(F)")
```

```python
# The mean and standard deviation of this posterior data points sample
# are the corollary of frequentist max likelihood estimate above
print(""" 
      F_true = {0}
      F_est = {1:.0f} +/- {2:.0f} (based on {3} measurements)
      """.format(F_true,
                 np.mean(sample), # sum(w * F) / sum(w),
                 np.std(sample), #sum(w)**-0.5,
                 N))
```

```python
# CONCLUSION: bayesian approach yields same result as frequentist.
```

```python
## VARYING PHOTON COUNTS: FREQUENTIST APPROACH
```

```python
# ADDING A DIMENSION: exploring the stochastic flux model
# Instead of assuming star flux is static, we now assume it has some
# stochastic variation (varies with time unpredictably)
# Model used: 2-parameter Gaussian: theta = [mu, sigma], where sigma =
# stdev of variability intrinsic to the object. 
Image(filename="images/F_true_stochastic.png")
```

```python
np.random.seed(42)
N = 100
mu_true, sigma_true = 1000, 15 # stochastic flux model

F_true = stats.norm(mu_true, sigma_true).rvs(N) # (unknown) true flux
F = stats.poisson(F_true).rvs() # observed flux: true flux + poisson errors.
e = np.sqrt(F) # errors formula

print("F_true = {0},\n\nF = {1}, \n\ne = {2}".format(F_true, F, e))
```

```python
# The likelihood of this model is the mix of the intrinsic distribution
# with the error distribution... (see formula)

Image(filename="images/likelihood2.png")
```

```python
# The parameters: need to maximize likelihood above to find the best
# estimate for the mean mu:
Image(filename="images/params.png")
```

```python
# But there is a problem: since optimal mu depends on optimal sigma, 
# there is correlation so cannot use analytic calculus. Must use
# numerical optimization. 
def logLikelihood(theta, F, e):
    mu = theta[0]
    sigma = theta[1]
    return -0.5*np.sum(np.log(2*np.pi * (sigma**2 + e**2)) + 
                       (F - mu)**2 / (sigma**2 + e**2))

# maximize lik <--> minimize neg lik
def negLogLikelihood(theta, F, e):
    return -logLikelihood(theta, F, e)


from scipy import optimize 
theta_guess = [900, 5]
theta_est = optimize.fmin(negLogLikelihood, theta_guess, args=(F, e))
print("""
        Maximum likelihood estimate for {0} data points: 
        mu = {theta[0]:.0f}, sigma={theta[1]:.0f}
""".format(N, theta=theta_est))
```

```python
# So parameters are computed - check
# But need to determine confidence in our answer (error bars)
    # 1. chi-square depends on strong assumptions of Gaussianity of 
    # likelihood.
    # 2. randomized sampling (bootstrap or jacknife) to maximize    
    # likelihood for randomized samples of the input data to 
    # explore degree of certainty in our result. 
from astroML.resample import bootstrap 

def fitSamples(sample):
    # sample = array of size [n_bootstraps, n_samples]
    # compute the maximum likelihood for each bootstrap. 
    return np.array([optimize.fmin(negLogLikelihood, theta_guess,
                                   args=(F, np.sqrt(F)), disp=0)
                     for F in sample])

# 1000 bootstrap resamplings
samples = bootstrap(F, 1000, fitSamples)
```

```python
mu_samp = samples[:, 0] # all rows, first col
sig_samp = abs(samples[:, 1]) # all rows second col

print("mu = {0:.0f} +/- {1:.0f}".format(mu_samp.mean(), 
                                        mu_samp.std()))
print("sigma = {0:.0f} +/- {1:.0f}".format(sig_samp.mean(), 
                                           sig_samp.std()))
```

```python
# NOTE: errors can potentially be correlated or non-Gaussian, not
# reflected here by simply finding mean and standard deviation
# of each model parameter.  But basic frequentist approach is shown.
```

```python
## VARYING PHOTON COUNTS: BAYESIAN APPROACH
```

```python
def logPrior(theta):
    # sigma needs to be positive.
    if theta[1] <= 0:
        return -np.inf 
    else:
        return 0 
    
def logPosterior(theta, F, e):
    return logPrior(theta) + logLikelihood(theta, F, e)

# same setup as above: 
ndim, nwalkers = 2, 50
nsteps, nburn = 2000, 1000

startingGuesses = np.random.rand(nwalkers, ndim)
startingGuesses[:, 0] *= 2000 # start mu between 0 and 2000
startingGuesses[:, 1] *= 20 # start sigma between 0 and 20

sampler = emcee.EnsembleSampler(nwalkers, ndim, logPosterior,
                                args=[F, e])
sampler.run_mcmc(startingGuesses, nsteps)

sample = sampler.chain # shape = (nwalkers, nsteps, ndim)
sample = sampler.chain[:, nburn:, :].reshape(-1, 2)
```

```python
# going to plot traces and contours representing one and two standard
# deviations. 

from astroML.plotting import plot_mcmc

fig = plt.figure()
plt.gcf().set_facecolor('white') # axes bg color 
ax = plot_mcmc(sample.T, fig=fig, labels=[r'$\mu$', r'$\sigma$'],
               colors='k')
ax[0].plot(sample[:, 0], sample[:, 1], ',k', alpha=0.1)
ax[0].plot([mu_true], [sigma_true], 'o', color='red', ms=10);

# red dot means the ground truth (from problem setup)
# contours mean one and two stddevs (68% and 95% conf levels).
# So 68% conf that model lies within inner contour and 95% conf that
# models lies within outer contour. 

# CONCLUSION: our data are not enough to confidently rule out the 
# possibility of a non-varying source of flux.
```

```python
# ANOTHER THING: posterior is not Gaussian - lack of symmetry in
# the vertical direction. This means Guassian approximation used
# in frequentist approach may not reflect the true uncertainties in the
# result. This results from implicit assumption of Gaussianity.
# In Bayesianism, this decision is explicit.
```
